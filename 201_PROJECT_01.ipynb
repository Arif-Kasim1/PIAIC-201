```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from google.colab import userdata
from langchain.prompts import PromptTemplate
from langchain.chains import SimpleSequentialChain, LLMChain, SequentialChain
# from langchain.memory import SimpleMemory

recipe_llm = HuggingFaceEndpoint(
    repo_id = "Qwen/QwQ-32B-Preview",
    huggingfacehub_api_token = userdata.get('HUGGINGFACE_API_KEY')
)

health_advisory_llm = HuggingFaceEndpoint(
    repo_id = "meta-llama/Llama-3.2-3B-Instruct",
    huggingfacehub_api_token = userdata.get('HUGGINGFACE_API_KEY')
)


recipe_template = PromptTemplate.from_template(
    "generate a recipe for a delicious {flavour} {item}")

health_advisory_templete = PromptTemplate.from_template(
    """calculate calorie and nutrition facts for the following dish
    and also recomand the safe intake amount for a Type 1 diabetic
    and type 2 diabetic patient separately for both type {recipe}""")

recipe_input_variables = {"flavour":"chocolate", "item":"cake"}
# user_recipe = recipe_template.format(**recipe_input_variables) #use ** to unpack dictionary
# print(user_recipe)

# result = llm.invoke(prompt)
#chat_modal = ChatHuggingFace(llm=llm)

# the output key is what gets passed to the next chain.
chain_1 = LLMChain(llm=recipe_llm, prompt=recipe_template, output_key="recipe")
chain_2 = LLMChain(llm=health_advisory_llm, prompt=health_advisory_templete,
                   output_key='health_advisory')

print(chain_1.output_keys)
print(chain_2.output_keys)

print(chain_1)
print(chain_2)


overall_chain = SequentialChain(chains=[chain_1, chain_2],
                                input_variables = ['flavour','item'],
                                output_variables=['recipe', 'health_advisory'])

chain_response = overall_chain.invoke(recipe_input_variables)

print(chain_response)

print(chain_response.get('recipe'))
print("\n\n *** Health advisory for the above recipe ***")
print(chain_response.get('health_advisory'))
# print(result)
# print(chat_modal.invoke(user_recipe))
```
