```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.schema import AIMessage, HumanMessage, SystemMessage
from google.colab import userdata
import os
import textwrap

os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_API_KEY')
```

```python
llm = ChatGoogleGenerativeAI(
              model = "gemini-2.0-flash-exp",
              temperature = 0.7,
              api_key = userdata.get('GOOGLE_API_KEY') )

promt_templet = ChatPromptTemplate.from_messages(
    [
        MessagesPlaceholder(variable_name="history1")

        # HumanMessage(content="Which laptop you suggest for home use.")
    ]
)

history: str = [
            SystemMessage(content="Your computer laptop expert."),
            SystemMessage(content="Dont entertain if question is not related to computer laptop.")
          ]

promt_templet.format(history1=history)
all_messages = history

while True:
  user_input = input("You: ")
  if user_input == 'exit':
    break
  # print(user_input)
  all_messages.append(HumanMessage(content=user_input))

  prompt = promt_templet.format(history1=all_messages)

  print("Prompt: ", prompt)

  response = llm.invoke(prompt)
  print("Model_Response: ", textwrap.wrap(response.content, width=80))
  all_messages.append(AIMessage(content=response.content))


```
